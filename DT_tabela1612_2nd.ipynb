### SOYBEAN IBGE 1612 DATA from 2000 to 2016

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
areapc = pd.read_excel("../tabular/period_soja_all.xlsx",sheetname='perc_area_soja',index_col='cod')
areatotha = pd.read_excel("../tabular/period_soja_all.xlsx",sheetname='area_total_ha',index_col='cod')
areasojaha = pd.read_excel("../tabular/period_soja_all.xlsx",sheetname='area_soja_ha',index_col='cod')
rendkgha = pd.read_excel("../tabular/period_soja_all.xlsx",sheetname='rend_soja_kg_ha',index_col='cod')
quantton = pd.read_excel("../tabular/period_soja_all.xlsx",sheetname='quant_soja_ton',index_col='cod')
```

```python
allmedians = pd.DataFrame(index=areapc.index)
```

```python
allmedians['areapc'] = areapc.apply(np.nanmedian,axis=1)
allmedians['areatotha'] = areatotha.apply(np.nanmedian,axis=1)
allmedians['areasojaha'] = areasojaha.apply(np.nanmedian,axis=1)
allmedians['rendkgha'] = rendkgha.apply(np.nanmedian,axis=1)
allmedians['quantton'] = quantton.apply(np.nanmedian,axis=1)
allmedians['important_areas'] = allmedians['rendkgha'] * allmedians['areapc']
```

```python
allmedians.to_csv("../tabular/p_s_medians.csv")
```

### 5-year mean total area
* this is to see how the soybean areas expanded in 5-year steps
* this will also help me to decide about data extraction
* I have to plot these differences in the map
    * If expansion is detected, I could either use the oldest crop mask (around 2002)
    * OR use more masks, yearly (or 3, or 5-year) based to extract data

```python
area5y = pd.DataFrame(index=areatotha.index)
area5y['2002-2006'] = areatotha.iloc[:,2:7].apply(np.nanmean,axis=1)
area5y['2007-2011'] = areatotha.iloc[:,7:12].apply(np.nanmean,axis=1)
area5y['2012-2016'] = areatotha.iloc[:,12:17].apply(np.nanmean,axis=1)
```
