```python
#DAY
#MOD and MYD folders
ft = '/media/denis/seagate/BRASIL/MODIS/MOD11A2/h13v11/day/'
fa = '/media/denis/seagate/BRASIL/MODIS/MYD11A2/h13v11/day/'

lt = xu.listfiles(ft,'*.tif',full=False)
la = xu.listfiles(fa,'*.tif',full=False)
lc = sorted(list(set(lt).intersection(la)))

#create dir for day
f = '/media/denis/seagate/BRASIL/MODIS/MxD11A2/h13v11/day/'
os.makedirs(f)

for i in lc:
    mod,m = xu.TifToArray(ft+i)
    myd,m = xu.TifToArray(fa+i)
    mxd = np.where(myd==0,mod+day[i.split('-')[1]]/0.02,myd)
    mxd = np.where(mxd<400,0,mxd)
    xu.ArrayToTif(mxd,i,f,m,1)
```

```python
#NIGHT
#MOD and MYD folders
ft = '/media/denis/seagate/BRASIL/MODIS/MOD11A2/h13v11/night/'
fa = '/media/denis/seagate/BRASIL/MODIS/MYD11A2/h13v11/night/'

lt = xu.listfiles(ft,'*.tif',full=False)
la = xu.listfiles(fa,'*.tif',full=False)
lc = sorted(list(set(lt).intersection(la)))

#create dir for night
f = '/media/denis/seagate/BRASIL/MODIS/MxD11A2/h13v11/night/'
os.makedirs(f)

for i in lc:
    mod,m = xu.TifToArray(ft+i)
    myd,m = xu.TifToArray(fa+i)
    mxd = np.where(myd==0,mod+night[i.split('-')[1]]/0.02,myd)
    mxd = np.where(mxd<400,0,mxd)
    xu.ArrayToTif(mxd,i,f,m,1)
```
